# Data Exploration

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('../Datasets/weatherAUS.csv')
df.head(10)
```

```python
df.info()
```

```python
df.dropna(subset=['RainToday', 'RainTomorrow'], inplace=True)
df.info()
```

```python
df.describe()
```

#  Train/Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.3, random_state=42)
train, val = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train.shape, val.shape, test.shape
```

```python
year = pd.to_datetime(df.Date).dt.year

train = df[year<2015]
val = df[year==2015]
test = df[year>2015]
```

```python
copy = train
train_Y = copy.pop('RainTomorrow')
train_X = copy
```

```python
copy = val
val_Y = copy.pop('RainTomorrow')
val_X = copy
```

```python
copy = test
test_Y = copy.pop('RainTomorrow')
test_X = copy
```

# Data Cleaning & PreProcessing

```python
import numpy as np
```

```python
for X in [train_X, val_X, test_X]:
    X.pop('Date')
```

```python
num_train = train_X.select_dtypes(include=np.number).columns.to_list()
cat_train = train_X.select_dtypes('object').columns.to_list()
```

```python
num_train
```

```python
train_X[num_train].describe()
```

```python
df[cat_train].nunique(), train_X[cat_train].nunique()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
df[num_train].isna().sum()
```

```python
imputer.fit(df[num_train])
```

```python
train_X[num_train] = imputer.transform(train_X[num_train])
val_X[num_train] = imputer.transform(val_X[num_train])
test_X[num_train] = imputer.transform(test_X[num_train])
```

```python
train_X[num_train].info()
```

```python
train_X[num_train].describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df[num_train])
```

```python
train_X[num_train] = scaler.transform(train_X[num_train])
val_X[num_train] = scaler.transform(val_X[num_train])
test_X[num_train] = scaler.transform(test_X[num_train])
```

```python
train_X.describe()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoder.fit(df[cat_train])
```

```python
encoded_cols = list(encoder.get_feature_names(cat_train))
encoded_cols
```

```python
encoder.get_feature_names(cat_train)
```

```python
train_X[encoded_cols] = encoder.transform(train_X[cat_train])
val_X[encoded_cols] = encoder.transform(val_X[cat_train])
test_X[encoded_cols] = encoder.transform(test_X[cat_train])
```

```python
for X in [train_X, val_X, test_X]:
    for feature in cat_train:
        X.pop(feature)
```

 # Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear')
model.fit(train_X, train_Y)
```

```python
weights_df = pd.DataFrame({
    'Features': (train_X.columns),
    'Weights': model.coef_[0]
})
weights_df
```

```python
train_Y.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,50))
sns.barplot(data = weights_df, x='Weights', y='Features')
```

```python
val_preds = model.predict(val_X)
test_preds = model.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

val_acc = accuracy_score(val_preds, val_Y) 
test_acc = accuracy_score(test_preds, test_Y) 

print('Accuracies: ', val_acc, ',', test_acc)
```

```python
matrix = confusion_matrix(test_Y, test_preds, normalize='true')
```

```python
sns.heatmap(matrix)
```

# Save & Load

```python
import joblib
```

```python
aus_rain = {
    'model': model,
    'imputer': imputer,
    'scaler': scaler,
    'encoder': encoder,
    'train_X': train_X,
    'train_Y': train_Y,
    'num_cols': num_train,
    'cat_cols': cat_train,
    'encoded_cols': encoded_cols
}
```

```python
joblib.dump(aus_rain, '../models/logistic_ausRain.joblib')
```
